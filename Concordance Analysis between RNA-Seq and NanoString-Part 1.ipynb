# 2- Bland-Altman

```python
import seaborn as sns

# Set the color palette
sns.set_palette('pastel')

# Iterate over the rows (conditions) of the dataframes
for condition in NanoString.index:
    # Get the gene expression values for the current condition
    NanoString_normalized_vals = NanoString.loc[condition, :]
    RNASeq_normalized_vals = RNASeq.loc[condition, :]
    
    # Calculate the difference and mean values
    difference = NanoString_normalized_vals - RNASeq_normalized_vals
    mean_values = (NanoString_normalized_vals + RNASeq_normalized_vals) / 2
    
    # Calculate the mean difference and limits of agreement
    mean_difference = difference.mean()
    limits_of_agreement = 1.96 * difference.std()
    
    # Count genes within and outside limits of agreement
    total_genes = len(difference)
    within_limits = np.sum((difference <= (mean_difference + limits_of_agreement)) & (difference >= (mean_difference - limits_of_agreement)))
    outside_limits = total_genes - within_limits
    
    # Create the Bland-Altman plot for the current condition
    fig, ax = plt.subplots(figsize=(11, 4), dpi=100)
    
    # Scatter plot with custom markers and colors
    plt.scatter(mean_values, difference, alpha=0.7, marker='o', edgecolors='black', linewidths=0.7)
    
    # Add horizontal lines for the mean difference and limits of agreement
    plt.axhline(mean_difference, color='r', linestyle='--', linewidth=1.5)
    plt.axhline(mean_difference + limits_of_agreement, color='gray', linestyle='--', linewidth=1.0)
    plt.axhline(mean_difference - limits_of_agreement, color='gray', linestyle='--', linewidth=1.0)
    
    # Add annotations for the mean difference and limits of agreement
    for xy, text in [((plt.xlim()[1], mean_difference), f'Mean: {mean_difference:.3f}'),
                     ((plt.xlim()[1], mean_difference + limits_of_agreement), f'Upper Limit: {mean_difference + limits_of_agreement:.3f}'),
                     ((plt.xlim()[1], mean_difference - limits_of_agreement), f'Lower Limit: {mean_difference - limits_of_agreement:.3f}')]:
        plt.annotate(text, xy=xy, xytext=(10, 0), textcoords='offset points', fontsize=8)
    
    # Identify genes that are outside the limits of agreement
    outside_limits_genes = difference[(difference > (mean_difference + limits_of_agreement)) | (difference < (mean_difference - limits_of_agreement))]
    for gene, diff in outside_limits_genes.iteritems():
        plt.scatter(mean_values[gene], difference[gene], color='red', alpha=0.8, marker='o', edgecolors='black', linewidths=0.5, label=f'{gene}: {diff:.3f}')
    
    # Add labels and title
    plt.xlabel('Mean of NanoString_normalized and RNASeq_normalized', fontsize=12)
    plt.ylabel('Difference (NanoString- RNASeq)', fontsize=12)
    plt.title(f'Bland-Altman Plot - Condition: {condition} \n Within limits: {within_limits} ({within_limits/total_genes*100:.1f}%), Outside limits: {outside_limits} ({outside_limits/total_genes*100:.1f}%)', fontsize=16, fontweight='bold')
    
    # Customize plot aesthetics
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True, linestyle='--', linewidth=0.5)
    
    # Add legends in two columns
    handles, labels = ax.get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    legend = ax.legend(by_label.values(), by_label.keys(), title='Genes Outside Limits of Agreement \n          (Gene: Difference)', fontsize=8.5, ncol=5)

    # Set the fontsize of the legend title
    legend.get_title().set_fontsize('11')

    # Show the Bland-Altman plot for the current condition
    plt.tight_layout()
    plt.show()
```
